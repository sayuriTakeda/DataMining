```python
import pylab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("/home/sayuri/Documentos/iris_treino.csv")
```

```python
plt.plot(base.SepalLengthCm,label="comp. da spala",color="k")
plt.plot(base.SepalWidthCm,label="largura da spala",color='y')
plt.plot(base.PetalWidthCm,label="largura da ptala",color='teal')
plt.plot(base.PetalLengthCm,label="comp. da ptala",color='#7a5b99')
plt.ylabel('features')
plt.xlabel('cm')
plt.legend(loc=2,fontsize="small")
plt.ylim(0,10)
plt.show()
```

```python
f, axarr = plt.subplots(4, sharex=True)
axarr[0].plot(base.SepalLengthCm)
axarr[1].plot(base.SepalWidthCm)
axarr[2].plot(base.PetalLengthCm)
axarr[3].plot(base.PetalWidthCm)
plt.show()
```

```python
plt.plot(base.PetalLengthCm,color="#7a5b99")
plt.plot(base.PetalWidthCm,color='teal')
plt.legend(loc=2,fontsize="small")
plt.title('Ptala')
plt.xlabel('cm')
plt.show()
```

```python
plt.hist(base.PetalLengthCm,color='#7a5b99')
plt.hist(base.PetalWidthCm,facecolor='teal')
plt.annotate('provavelmente 2 espcies diferentes onde',horizontalalignment='left', xy=(2, 30), xytext=(2.5, 35),
            arrowprops=dict(facecolor='w', shrink=0.05)
            )
plt.annotate('o comprimento de uma  igual a largura de outra',xy=(2, 30), xytext=(2.5, 32))
plt.show()
```

```python
all_data = [base.PetalLengthCm,base.PetalWidthCm]

plt.boxplot(all_data,notch=True)            

plt.show()
```

```python
x = base.PetalLengthCm
y = base.PetalWidthCm

plt.scatter(x, y, c='lavenderblush', alpha=0.8)
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for c, m, zl, zh in [('r', 'o', -40, -25), ('b', '^', -30, -5)]:
    xs = base.PetalLengthCm
    ys = base.PetalWidthCm
    zs = base.SepalWidthCm
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
