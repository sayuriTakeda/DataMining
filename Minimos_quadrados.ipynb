```python
M = pd.read_csv("/home/sayuri/Área de Trabalho/Algelin/Pontos.csv",header=0)
M = np.matrix(M)
```

```python
Y = M[:,1]
```

```python
coluna1 = []
for i in range(128):
    coluna1.append(1)
```

```python
M = M[:,0]
```

```python
M = pd.DataFrame(M)
M['labels'] = coluna1
M = np.around(M,decimals=2)
M
```

```python
M = np.asmatrix(M)
```

```python
pseudo_inversa = np.linalg.pinv(M) # np.linalg.inv(M.T*M)*M.T
```

```python
M_pseu = M*pseudo_inversa
```

```python
linha = np.dot(M_pseu,Y)
```

```python
pontos = pd.DataFrame(pontos)
pontos['labels'] = linha
pontos = np.around(pontos,decimals=2)
pontos
```

```python
plt.plot(list(range(len(pontos))),pontos)
plt.show()
```
